Defining core functions:
* `get_sin(x)`: Returns the sine of x.
* `get_cos(x)`: Returns the cosine of x.

```python
import numpy as np
def get_sin(x):
    return np.sin(x)
def get_cos(x):
    return np.cos(x)  
```

```python
x_values = np.linspace(0, 2, 1000)

sin_values = get_sin(x_values)
cos_values = get_cos(x_values)
```

```python
print(f"{'x':>10} | {'sin(x)':>10} | {'cos(x)':>10}")
print("-" * 36)

for i in range(10):
    print(f"{x_values[i]:10.6f} | {sin_values[i]:10.6f} | {cos_values[i]:10.6f}")
```
